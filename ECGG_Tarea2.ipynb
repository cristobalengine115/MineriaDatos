* 24) famrel: Calidad de la relación del estudiante y su familia.

### **Paso 3: Detección de valores atípicos**

**1) Distribución de variables numéricas**

```python
DatosEstudiantes.hist(figsize=(14,14), xrot=45)
plt.show()
```

**Resumen estadístico de variables numéricas**

```python
DatosEstudiantes.describe()
```

**Diagramas para detectar posibles valores atípicos**

```python
VariablesValoresAtipicos = ['failures', 'traveltime', 'Dalc', 'absences']
for col in VariablesValoresAtipicos:
  sns.boxplot(data=DatosEstudiantes[col],orient='h').set_title(col)
  plt.show()
```

**Distribución de variables categóricas**

```python
DatosEstudiantes.describe(include='object')
```

```python
for col in DatosEstudiantes.select_dtypes(include='object'):
  if DatosEstudiantes[col].nunique()<10:sns.countplot(y=col, data=DatosEstudiantes)
  plt.show()
```

**Arupación por variables categóricas**

```python
for col in DatosEstudiantes.select_dtypes(include='object'):
  if DatosEstudiantes[col].nunique() < 10:
    display(DatosEstudiantes.groupby(col).agg(['mean']))
```

### Paso 4: **Identificación de relaciones entre pares variables**

```python
DatosEstudiantes.corr()
```

```python
plt.figure(figsize=(14,7))
sns.heatmap(DatosEstudiantes.corr(), cmap='RdBu_r', annot=True)
plt.show()
```

```python
plt.figure(figsize=(14,7))
MatrizInf = np.triu(DatosEstudiantes.corr())
sns.heatmap(DatosEstudiantes.corr(), cmap='RdBu_r', annot=True, mask=MatrizInf)
plt.show()
```
